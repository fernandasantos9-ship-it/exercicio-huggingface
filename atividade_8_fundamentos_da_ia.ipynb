```

!pip install transformers torch

from transformers import pipeline


modelo_1 = pipeline('text-generation', model='gpt2')
modelo_2 = pipeline('text-generation', model='distilgpt2')

prompt = "The future of Artificial Intelligence is"


print("--- Gerando com GPT-2 ---")
res1 = modelo_1(prompt, max_length=30, num_return_sequences=1)
print(res1[0]['generated_text'])


print("\n--- Gerando com DistilGPT2 ---")
res2 = modelo_2(prompt, max_length=30, num_return_sequences=1)
print(res2[0]['generated_text'])
```
